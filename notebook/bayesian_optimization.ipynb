```python
from skopt import gp_minimize
from skopt.space import Real
from skopt.utils import use_named_args
from skopt.learning import GaussianProcessRegressor
from skopt.callbacks import EarlyStopper
import datetime
from sklearn.gaussian_process.kernels import RBF, Matern, WhiteKernel, ConstantKernel as C
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import os
```

#**ã‚«ãƒ¼ãƒãƒ«é–¢æ•°ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆ - åˆæœŸå€¤ã ã‘25,5ã§ãã‚Œä»¥é™ã‹ã‚‰æœ€é©åŒ–ãƒ»é›†ä¸­åº¦9.8ä»¥ä¸Šã§çµ‚äº†**

```python
# ğŸ”§ æ¢ç´¢ç©ºé–“ã®å®šç¾©
space = [
    Real(15, 60, name="work_duration"),    # ä½œæ¥­æ™‚é–“ï¼ˆåˆ†ï¼‰
    Real(3, 20, name="break_duration"),    # ä¼‘æ†©æ™‚é–“ï¼ˆåˆ†ï¼‰
]

# ğŸ“¦ ãƒ­ã‚°ãƒ•ã‚¡ã‚¤ãƒ«ã®åˆæœŸåŒ–
log_file = "pomodoro_optimization_log.csv"
if not os.path.exists(log_file):
    pd.DataFrame(columns=["datetime", "work_duration", "break_duration", "focus_score"]).to_csv(log_file, index=False)

def simulate_focus_score(work_duration, break_duration):
    # ç†æƒ³å€¤ã¨ã®å·®ã‚’å–ã£ã¦ã‚¹ã‚³ã‚¢ã‚’ä¸‹ã’ã‚‹ï¼ˆç†æƒ³ï¼šä½œæ¥­30åˆ†ã€ä¼‘æ†©7åˆ†ï¼‰
    work_penalty = abs(work_duration - 30) / 6     # Â±6åˆ†ãšã‚Œã‚‹ã”ã¨ã«1ç‚¹æ¸›ç‚¹
    break_penalty = abs(break_duration - 7) / 2.5  # Â±2.5åˆ†ãšã‚Œã‚‹ã”ã¨ã«1ç‚¹æ¸›ç‚¹

    base_score = 10 - (work_penalty + break_penalty)

    # ãƒã‚¤ã‚ºã‚’è¿½åŠ ï¼ˆÂ±0.5ç¨‹åº¦ã®ãƒ–ãƒ¬ï¼‰
    noise = np.random.normal(0, 0.5)
    score = base_score + noise

    # ã‚¹ã‚³ã‚¢ã¯1ã€œ10ã«ã‚¯ãƒªãƒƒãƒ—
    return float(np.clip(score, 1, 10))


# ğŸ§  æœ€é©åŒ–ã™ã‚‹ç›®çš„é–¢æ•°
@use_named_args(space)
def objective(when=None, **params):
    print(f"\nğŸ•’ ãƒãƒ¢ãƒ‰ãƒ¼ãƒ­è¨­å®šã‚’ææ¡ˆã—ã¾ã™:")
    print(f"ğŸ‘‰ ä½œæ¥­æ™‚é–“: {params['work_duration']:.1f}åˆ†")
    print(f"ğŸ‘‰ ä¼‘æ†©æ™‚é–“: {params['break_duration']:.1f}åˆ†")

    # ğŸ”€ æ“¬ä¼¼çš„ã«é›†ä¸­åº¦ã‚’ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆ
    score = simulate_focus_score(params["work_duration"], params["break_duration"])
    print(f"âœ¨ è‡ªå‹•ç”Ÿæˆã•ã‚ŒãŸé›†ä¸­åº¦ã‚¹ã‚³ã‚¢: {score:.2f}")
    if when is None:
      when = datetime.now()
    print(f"âœ¨ å®Œäº†ã—ãŸæ™‚é–“: {when}")

    # ãƒ­ã‚°ã«ä¿å­˜
    new_entry = {
        "datetime": datetime.now(),
        "work_duration": params["work_duration"],
        "break_duration": params["break_duration"],
        "focus_score": score
    }
    df = pd.read_csv(log_file)
    df = pd.concat([df, pd.DataFrame([new_entry])], ignore_index=True)
    df.to_csv(log_file, index=False)

    return -score

TARGET_SCORE = 9.8

# ã‚³ãƒ¼ãƒ«ãƒãƒƒã‚¯é–¢æ•°
def call_back(res):
    current_best = -res.fun
    if current_best >= TARGET_SCORE:
      return True
    return False

initial_point = [25, 5]
initial_score = simulate_focus_score(25, 5)

# ğŸ” æœ€é©åŒ–ã®å®Ÿè¡Œ
result = gp_minimize(
    func=objective,
    dimensions=space,
    n_calls=1000,
    n_initial_points=1,
    x0=[initial_point],
    y0=[initial_score],
    random_state=42,
    verbose=True,
    callback=[call_back]
)

# ğŸ“Š çµæœã®å¯è¦–åŒ–
print("\nâœ… æœ€é©ãªè¨­å®šï¼ˆé›†ä¸­åº¦æœ€å¤§ï¼‰:")
print(f"ä½œæ¥­æ™‚é–“: {result.x[0]:.1f}åˆ†")
print(f"ä¼‘æ†©æ™‚é–“: {result.x[1]:.1f}åˆ†")
print(f"æœ€å¤§é›†ä¸­åº¦: {-result.fun:.1f}")

# åæŸãƒ—ãƒ­ãƒƒãƒˆ
from skopt.plots import plot_convergence
plot_convergence(result)
plt.show()
```

```python
# ğŸ”§ æ¢ç´¢ç©ºé–“ã®å®šç¾©
space = [
    Real(15, 60, name="work_duration"),    # ä½œæ¥­æ™‚é–“ï¼ˆåˆ†ï¼‰
    Real(3, 20, name="break_duration"),    # ä¼‘æ†©æ™‚é–“ï¼ˆåˆ†ï¼‰
]

# ğŸ“¦ ãƒ­ã‚°ãƒ•ã‚¡ã‚¤ãƒ«ã®åˆæœŸåŒ–
log_file = "pomodoro_optimization_log.csv"
if not os.path.exists(log_file):
    pd.DataFrame(columns=["datetime", "work_duration", "break_duration", "focus_score"]).to_csv(log_file, index=False)

# ğŸ¯ é›†ä¸­åº¦ã‚¹ã‚³ã‚¢ã®ãƒªã‚¢ãƒ«ãªãƒ¢ãƒ‡ãƒ«ï¼ˆç†æƒ³ã‹ã‚‰ã®ã‚ºãƒ¬ + ãƒã‚¤ã‚ºï¼‰
def simulate_focus_score(work_duration, break_duration):
    work_penalty = abs(work_duration - 30) / 6
    break_penalty = abs(break_duration - 7) / 2.5
    base_score = 10 - (work_penalty + break_penalty)
    noise = np.random.normal(0, 0.5)
    return float(np.clip(base_score + noise, 1, 10))

# ğŸ§  ç›®çš„é–¢æ•°
@use_named_args(space)
def objective(**params):
    print(f"\nğŸ•’ ææ¡ˆã•ã‚ŒãŸãƒãƒ¢ãƒ‰ãƒ¼ãƒ­è¨­å®š:")
    print(f"ğŸ‘‰ ä½œæ¥­æ™‚é–“: {params['work_duration']:.1f}åˆ†")
    print(f"ğŸ‘‰ ä¼‘æ†©æ™‚é–“: {params['break_duration']:.1f}åˆ†")
    score = simulate_focus_score(params["work_duration"], params["break_duration"])
    print(f"âœ¨ é›†ä¸­åº¦ã‚¹ã‚³ã‚¢ï¼ˆè‡ªå‹•ç”Ÿæˆï¼‰: {score:.2f}")

    # ãƒ­ã‚°ä¿å­˜
    new_entry = {
        "datetime": datetime.datetime.now(),
        "work_duration": params["work_duration"],
        "break_duration": params["break_duration"],
        "focus_score": score
    }
    df = pd.read_csv(log_file)
    df = pd.concat([df, pd.DataFrame([new_entry])], ignore_index=True)
    df.to_csv(log_file, index=False)

    return -score  # ã‚¹ã‚³ã‚¢æœ€å¤§åŒ– â†’ æœ€å°åŒ–å•é¡Œã«ã™ã‚‹ãŸã‚ç¬¦å·åè»¢

# ğŸ”§ ã‚«ã‚¹ã‚¿ãƒ åˆæˆã‚«ãƒ¼ãƒãƒ«ï¼ˆæ»‘ã‚‰ã‹ + ç²—ã• + ãƒã‚¤ã‚º + ãƒã‚¤ã‚¢ã‚¹ï¼‰
kernel = (RBF(length_scale=10.0) * Matern(length_scale=5.0, nu=1.5)) + WhiteKernel(noise_level=0.5) + C(1.0)

# ğŸ“ˆ GPRãƒ¢ãƒ‡ãƒ«ã«ã‚«ã‚¹ã‚¿ãƒ ã‚«ãƒ¼ãƒãƒ«ã‚’çµ„ã¿è¾¼ã‚€
base_estimator = GaussianProcessRegressor(kernel=kernel, normalize_y=True, random_state=42)

# ğŸ” ãƒ™ã‚¤ã‚ºæœ€é©åŒ–ã®å®Ÿè¡Œ
from skopt import gp_minimize
result = gp_minimize(
    func=objective,
    dimensions=space,
    base_estimator=base_estimator,
    n_calls=50,
    n_initial_points=5,
    random_state=42,
    acq_optimizer="sampling",
    verbose=True
)

# ğŸ“Š æœ€é©çµæœã®è¡¨ç¤º
print("\nâœ… æœ€é©ãªãƒãƒ¢ãƒ‰ãƒ¼ãƒ­è¨­å®š:")
print(f"ä½œæ¥­æ™‚é–“: {result.x[0]:.1f}åˆ†")
print(f"ä¼‘æ†©æ™‚é–“: {result.x[1]:.1f}åˆ†")
print(f"æœ€å¤§é›†ä¸­åº¦: {-result.fun:.2f}")

# ğŸ” åæŸãƒ—ãƒ­ãƒƒãƒˆ
from skopt.plots import plot_convergence
plot_convergence(result)
plt.tight_layout()
plt.show()
```
